#### Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action = 'ignore')
```

#### Importing the Dataset

```python
dataset = pd.read_csv('train.csv',usecols=['GarageQual','FireplaceQu','SalePrice'])
```

```python
dataset.head()
```

```python
#calculates the percentage of missing values in each column of the dataset DataFrame.
dataset.isnull().mean()*100
```

#### Visualising column before Imputation

```python
plt.figure(figsize = (5,3.5))
dataset['GarageQual'].value_counts().sort_values(ascending=False).plot.bar()
plt.xlabel('GarageQual')
plt.ylabel('Number of houses')
plt.title("GarageQual")
```

### Without Libraries

#### Imputing missing values using Missing Category Imputation

```python
dataset['GarageQual'].fillna('Missing', inplace=True)
```

#### Visualising column after Imputation

```python
plt.figure(figsize = (5,3.5))
dataset['GarageQual'].value_counts().sort_values(ascending=False).plot.bar()
plt.xlabel('GarageQual')
plt.ylabel('Number of houses')
plt.title("GarageQual")
```

### Using Libraries

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(dataset.drop(columns=['SalePrice']), dataset['SalePrice'],test_size=0.2)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='constant',fill_value='Missing')
```

```python
x_train = imputer.fit_transform(x_train)
x_test = imputer.transform(x_train)
```
