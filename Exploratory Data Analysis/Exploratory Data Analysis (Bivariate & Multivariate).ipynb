#### Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
warnings.filterwarnings(action = 'ignore')
```

#### Importing the Dataset

```python
titanic = pd.read_csv("Titanic.csv")
titanic.head()
```

```python
tips = pd.read_csv('Tips.csv')
tips.head()
```

```python
iris = pd.read_csv('Iris.csv')
iris.head()
```

#### 1. SCATTERPLOT (Numerical - Numerical)

```python
sns.set_style("dark")
sns.set(rc={'figure.figsize':(7.5,5)})
figure = sns.scatterplot(tips['total_bill'], tips['tip'], hue = tips['sex'], style = tips['smoker'], size = tips['size'], s = 20)
figure.set(xlabel = 'Total Bill',ylabel = 'Tip')
```

#### 2. BARPLOT (Numerical - Categorical)

```python
sns.barplot(titanic['Pclass'], titanic['Age'], hue = titanic['Survived'])
```

#### 3. BOXPLOT (Numerical - Catgerocial)

```python
sns.boxplot(tips['sex'], tips['tip'], hue = tips['smoker'])
```

#### 4. DISTPLOT (Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'], hist = False)
sns.distplot(titanic[titanic['Survived'] == 1]['Age'], hist = False)
```

#### 5. HEATMAP (Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

#### 6. CLUSTERMAP (Categorical - Categorical)

```python
sns.set(rc={'figure.figsize':(3.5,5)})
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```

```python
7. PAIRPLOT (All Numerical V/S All Numerical)
```

```python
sns.pairplot(iris, hue = 'species')
```
