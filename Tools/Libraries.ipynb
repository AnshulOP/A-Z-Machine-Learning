# Important libraries used in Machine Learning

## 1. Pandas

```python
# Used for data analysis and associated manipulation of tabular data in DataFrames.
import pandas as pd
```

```python
Raw_Housing_Data = pandas.read_csv("1. Regression - Module - (Housing Prices).csv")
Raw_Housing_Data
```

## Numpy

```python
# Used to solve large mathematical calculations in seconds with the help of saved presets.
import numpy as np
```

```python
np.mean(Raw_Housing_Data['Sale Price'])
```

## Matplolib

```python
# Used for ploting basic graphs used for data visualisation.
import matplotlib.pyplot as plt
```

```python
plt.plot(Raw_Housing_Data['Sale Price'].head(100),marker = 'o',markerfacecolor = 'Red',markersize = 5,linewidth = 5)
plt.xlabel("Record Number")
plt.ylabel("Sales Price")
plt.title("Sales Price Graph for Housing Data")
plt.show()
```
