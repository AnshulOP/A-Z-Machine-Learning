#### Importing the Libraries

```python
import numpy as np
```

#### Making a new Dataset

```python
from sklearn.datasets import make_regression
x, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

#### Visualising the values inside the data

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (4.7,3.2))
plt.scatter(x, y)
```

#### Apllying OLS to get the values of one variable

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x, y)
```

```python
print('Slope :',regressor.coef_)
```

```python
print('Intercept :',regressor.intercept_)
```

#### Visualising the Regression Line

```python
plt.figure(figsize = (4.7,3.2))
plt.scatter(x, y)
plt.plot(x, regressor.predict(x), color='red')
```

### Applying Gradient Descent

```python
# Applying Gradient Descent assuming slope is constant m = 48.53
# And assuming the starting value for intercept b = 0
y_pred = ((78.35 * x) + 0).reshape(4)
```

#### At b = 0

```python
plt.figure(figsize = (7,4))
plt.scatter(x, y)
plt.plot(x, regressor.predict(x),color='red',label='OLS')
plt.plot(x, y_pred, color='blue', label='b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0

slope = -2 * np.sum(y - m*x.ravel() - b)
slope
```

```python
# Taking Learning Rate of 0.1
rate = 0.1

step_size = slope*rate
step_size
```

```python
# Calculating the new value of intercept
b = b - step_size
b
```

#### At b = 20.927763408209216

```python
y_pred1 = ((78.35 * x) + b).reshape(4)

plt.figure(figsize = (7,4))
plt.scatter(x, y)
plt.plot(x, regressor.predict(x), color='red', label='OLS')
plt.plot(x,y_pred, color='#A3E4D7', label='b = 0')
plt.plot(x,y_pred1, color='#00a65a', label='b = {}'.format(b))
plt.legend()
plt.show()
```

#### Funciton to apply Gradient Descent for any assumed value of b

```python
# Desired value of b is around 26
b = -50
m = 78.35
rate = 0.08

epochs = 40

for i in range(epochs):
  slope = -2 * np.sum(y - m*x.ravel() - b)
  b = b - (rate * slope)
  print('Value of b at interation', i+1, b)  
  y_pred = m * x + b

  plt.plot(x, y_pred)

plt.scatter(x, y)
     
```
